```python
def fib(number):
    if number == 1 or number == 0:
        return number
    else:
        return fib(number-1) + fib(number-2)
print(fib(0))
```

```python
arr = [1,2,3]
org = []
def gen_ss(index, sum, org):
    if index == len(arr) - 1:
        # print(sum)
        org.append(sum)
        return  
    gen_ss(index+1, sum + arr[index], org)
    gen_ss(index+1, sum + arr[index+1], org)

print(gen_ss(0, 0, org))
print(org)
```
